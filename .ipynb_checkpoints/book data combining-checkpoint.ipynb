```python
import pandas as pd
```

```python
df_1=pd.read_csv("Book.csv")
df_2=pd.read_csv("Details.csv")
```

```python
df_1.head()
```

```python
df_2.head()
```

### Combining unique column of tweo data set into one

#### Creating Empty dataset

```python
df=pd.DataFrame()
```

#### Adding columns in dataset

```python
df["Title"]          =df_1["Title"]
df["Category"]       =df_2["Category"]
df["UPC"]            =df_2["UPC"]
df["Ratings"]        =df_1["Ratings"]
df["Reviews"]        =df_2["Review"]
df["Availability"]   =df_1["Availability"]
df["Quantity"]       =df_2["Quantity"]

df["Price_exlc_tax"] =df_2["Price_exlc_tax"]
df["Price_incl_tax"] =df_2["Price_incl_tax"]
df["Tax"]            =df_2["Tax"]
df["Book_Link"]      =df_1["Book_link"]
df["Img_Link"]       =df_1["Img_link"]

```

### Converting dataset in CSV format

```python
df.head()
```

```python
df.to_csv("Books_detail.csv",index=False)
```
