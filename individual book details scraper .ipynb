```python
import pandas as pd
import requests
from tqdm import tqdm 
from bs4 import BeautifulSoup
```

```python
df= pd.read_csv("Book.csv")
```

### Approach 1

```python
data=[]
for link in tqdm(df["Book_link"]):
    res=requests.get(link)
    soup=BeautifulSoup(res.text,"html.parser")

    book_type          =soup.find("ul",class_="breadcrumb").find_all("a")[2].text
    upc                =soup.find("table",class_="table table-striped").find_all("td")[0].text
    price_x            =soup.find("table",class_="table table-striped").find_all("td")[2].text[2:]
    price_i            =soup.find("table",class_="table table-striped").find_all("td")[3].text[2:]
    tax                =soup.find("table",class_="table table-striped").find_all("td")[4].text[2:]
    qty                =soup.find("table",class_="table table-striped").find_all("td")[5].text
    review             =soup.find("table",class_="table table-striped").find_all("td")[6].text

    data.append([book_type,upc,price_x,price_i,tax,qty,review])
    

data
```

### Aproach 2

```python
data=[]
for link in tqdm(df["Book_link"]):
    res=requests.get(link)
    soup=BeautifulSoup(res.text,"html.parser")
    temp=soup.find("table",class_="table table-striped").find_all("td")
    book_type          =soup.find("ul",class_="breadcrumb").find_all("a")[2].text
    upc                =temp[0].text
    price_x            =temp[2].text[2:]
    price_i            =temp[3].text[2:]
    tax                =temp[4].text[2:]
    qty                =temp[5].text
    review             =temp[6].text

    data.append([book_type,upc,price_x,price_i,tax,qty,review])

data
    
```

```python
new_df=pd.DataFrame(data,columns=["Category","UPC","Price_exlc_tax","Price_incl_tax","Tax","Quantity","Review"])
new_df.head()
```

```python
new_df.to_csv("Details.csv",index=False)
```
